Questão 5)

```python
homem = df[df['Sexo'] == 'M']
homem
```

```python
mulher = df[df['Sexo'] == 'F']
mulher
```

```python
mediaM = homem['limite_credito'].mean()
mediaF = mulher['limite_credito'].mean()
desvioM = homem['limite_credito'].std()
desvioF = mulher['limite_credito'].std()

print('Media Homem', mediaM)
print('Media Mulher', mediaF)
print('Desvio Homem', desvioM)
print('Desvio Mulher', desvioF)
```

```python
import seaborn as sns
import scipy.stats as st

n1 = len(homem['limite_credito'])
n2 = len(mulher['limite_credito'])
print(n1)
print(n2)

difstd = np.sqrt((np.power(desvioM,2)/n1) + np.power(desvioF,2)/n2)
difMedia = mediaM - mediaF

print('Diferença: %.3f' %difMedia)
print('Erro padrao dif: %.3f' %difstd)
print("As diferenças de média está %.3f vezes o erro padrao da média" %(difMedia/difstd))

var = 0.05
valorp = (st.norm.cdf(difstd))
print('(normal) valor-o = %.3f' %valorp)

if(valorp<var):
  print('Hipotese nula pode ser rejeitada')
else:
  print('Não podemos rejeitar a hipotese nula')
```
