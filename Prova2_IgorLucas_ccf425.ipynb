Fazendo a Regrass√£o linear:

```python
from sklearn.linear_model import LinearRegression

def regressao_linear(X,Y,cols):

  if len(cols) == 1:
    X = X.values.reshape(-1,1)

  lm = LinearRegression()
  lm.fit(X,Y)
  print('Coeficiente estimado: ', lm.coef_)    
  print('R2 (score): ', lm.score(X,Y))
  print("Intercept %.3f " % lm.intercept_)

  coeff_df = pd.DataFrame(lm.coef_, cols, columns=['Coefficient'])
  print(coeff_df)

Y = df[target]
Y
```

abaixo, temos uma precis√£o maior das rela√ß√µes

```python
var = ['Sexo', 'renda','nivel_educacao', 'limite_credito']
target = df['categoria_cartao']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import cluster
```

```python
kmeans = cluster.KMeans(n_clusters=2).fit(df[var])
```

Analisando o kmeans sempre utilizando o limite de credito como X.

```python
def plot_cluster(model, df, attr_x, attr_y):
  plt.figure(figsize=(10,3))
  sns.scatterplot(x=df[attr_x], y=df[attr_y], hue=model.labels_, s=40)
  plt.title('Classifica√ß√£o K-Means') 
    
  plt.show()
```

```python
X = df['limite_credito']
```

```python
plt.subplot(1,2,1)
sns.scatterplot(x=df['limite_credito'], y=df['Sexo'], hue=target, s=40)
plt.title('Classifica√ß√£o Real')

plt.subplot(1,2,2)
sns.scatterplot(x=df['limite_credito'], y=df['Sexo'], hue=kmeans.labels_, s=40)
plt.title('Classifica√ß√£o K-Means') 

plt.show()
```

```python
plt.subplot(1,2,1)
sns.scatterplot(x=df['limite_credito'], y=df['renda'], hue=target, s=40)
plt.title('Classifica√ß√£o Real')

plt.subplot(1,2,2)
sns.scatterplot(x=df['limite_credito'], y=df['renda'], hue=kmeans.labels_, s=40)
plt.title('Classifica√ß√£o K-Means') 

plt.show()
```

```python
plt.subplot(1,2,1)
sns.scatterplot(x=df['limite_credito'], y=df['nivel_educacao'], hue=target, s=40)
plt.title('Classifica√ß√£o Real')

plt.subplot(1,2,2)
sns.scatterplot(x=df['limite_credito'], y=df['nivel_educacao'], hue=kmeans.labels_, s=40)
plt.title('Classifica√ß√£o K-Means') 

plt.show()
```

Agora vendo a silhueta entre eles:

```python
from sklearn import metrics
```

```python
print('Silhueta:', round(metrics.silhouette_score(df[var], kmeans.labels_), 4))
```

Com isso, percebe-se que os valores da nossa analise est√° 0,7595 com valores longes do outro cluster
